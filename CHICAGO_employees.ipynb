```python
chicago_employees['Annual Salary'] = chicago_employees['Annual Salary'].apply(lambda x: x.replace(',', '')
                                if isinstance(x, str) else x).astype(float)
```

```python
#Nan are the 'unknown'
chicago_employees_top_salaries=chicago_employees[['Job Titles', 'Annual Salary']].sort_values(by='Annual Salary', ascending=False)
chicago_employees_top_salaries.head()
```
