```python
blog_buy = pd.read_csv('pandas_tutorial_buy.csv', delimiter=';', names = ['my_date_time', 'event', 'user_id', 'amount'])
```

```python
blog_buy.sample(7)
```

#### task1: Whatâ€™s the average (mean) revenue between 2018-01-01 and 2018-01-07 from the users in the article_read dataframe?


```python
a_read_buy=article_read.merge(blog_buy, how = 'left', left_on = 'user_id', right_on = 'user_id')
```

```python
a_read_buy.sample(50)
```

```python
#I removed all the unnecessary columns, and kept only amount
b_read_buy=a_read_buy.amount
```

```python
b_read_buy
```

```python
#I replaced NaN values with 0s.
c_read_buy=b_read_buy.fillna(0)
c_read_buy
```

```python
result=c_read_buy.mean()
result
```

```python
step_1 = article_read.merge(blog_buy, how = 'left', left_on = 'user_id', right_on = 'user_id')
step_2 = step_1.amount
step_3 = step_2.fillna(0)
result = step_3.mean()
result
```
